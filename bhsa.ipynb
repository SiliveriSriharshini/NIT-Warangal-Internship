```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

rmse_values = []

def fitness_function(selected_features):
    original_dataframe = df1.copy()
    selected_feature_names = original_dataframe.columns[np.where(selected_features == 1)[0]]
    if selected_feature_names.size == 0:
        # Handle the case where no features are selected
        return float('inf')
    subset_dataframe = original_dataframe[selected_feature_names]
    x = subset_dataframe
    y = df['Effort']

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)

    RFregressor = RandomForestRegressor()
    RFregressor.fit(x_train, y_train)
    score = RFregressor.score(x_test, y_test)
    y_preds = RFregressor.predict(x_test)

    rmse = np.sqrt(metrics.mean_squared_error(y_test, y_preds))
    b = round(rmse, 2)
    return b
N=7    
for i in range(R):
    selected_features = np.array(dataset1.iloc[i, ])
    b = fitness_function(selected_features)
    rmse_values.append(b)

dataset['rmse'] = rmse_values

def BHSA(HMS, HMCR, PAR, BW, N):
    r1 = np.random.random()
    r2 = np.random.random(1)[0]
    r3 = np.random.random(1)[0]
    if r1 < HMCR:
        data = np.random.choice([0, 1], size=(1, N), p=[0.4, 0.6])
        dff = pd.DataFrame(data, columns=['Input', 'Output', 'Inquiry', 'File', 'FPAdj', 'RawFPcounts', 'AdjFP'])
        dff[dff.sum(axis=1) < N/2] = 1
        b = fitness_function(dff.iloc[0, :])
        if r2 < PAR:
            df_copy = np.array(dff)
            new_df = []
            for i in range(df_copy.shape[0]):  # Iterate over the rows
                for j in range(df_copy.shape[1]):
                    x = (df_copy[i,j]) + (r2 * BW)
                    z = 1/(1 + np.exp(-x))
                    if (r3 < z):
                        new_df.append(1)
                    else:
                        new_df.append(0)
            #c = fitness_function(new_df)
            if new_df:
                c = fitness_function(pd.DataFrame([new_df], columns=dff.columns))
            else:
                c = b  # Set an appropriate default value if fitness function cannot be calculated
            a = min(b, c)
            if a == b:
                new_harmony = dff
            else:
                new_harmony = new_df
            new_harmony['rmse']=a
        else:
            a = b  # Assign a default value when r2 >= PAR
            new_harmony = dff
            new_harmony['rmse']=a
    else:
        # Assign a default value when r1 >= HMCR
        a,new_harmony = BHSA(HMS, HMCR, PAR, BW, N)
    return a, new_harmony

def updating_data(dataset):
    a, new_harmony = BHSA(HMS, HMCR, PAR, BW,N)
    if a < dataset['rmse'].max():
        #print(new_harmony)
        worst_index = dataset['rmse'].idxmax()
        dataset.iloc[worst_index] = new_harmony
    else:
        a, new_harmony = BHSA(HMS, HMCR, PAR, BW, N)
        updating_data(dataset)
        
def find_best_worst_solution(dataset):
    best_index = dataset['rmse'].idxmin()
    worst_index = dataset['rmse'].idxmax()
    return best_index, worst_index
# Example usage
HMS = 10
max_iter = 10
HMCR = 0.8
PAR = 0.3
BW = 0.1
N=7

for i in range(max_iter):
    rmse, harmony = BHSA(HMS, HMCR, PAR, BW, N)
    updating_data(dataset)
best_index, worst_index = find_best_worst_solution(dataset)
optimal_solution=np.array(dataset.iloc[best_index])
print("optimal solution:", optimal_solution)
print("optimal rmse:", dataset['rmse'].min())
print("worst solution:", np.array(dataset.iloc[worst_index]))
print("worst rmse:", dataset['rmse'].max())
dataset
```

```python
import matplotlib.pyplot as plt

def plot_features_vs_rmse(dataframe):
    features = dataframe.columns[:-1]  # Exclude the 'rmse' column
    rmse_values = dataframe['rmse']

    plt.figure(figsize=(10, 6))
    for feature in features:
        plt.scatter(dataframe[feature], rmse_values, label=feature)
    plt.xlabel('Feature Values')
    plt.ylabel('RMSE')
    plt.title('Effect of Features on RMSE')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_features_vs_rmse(dataset)
```

```python
import seaborn as sns
def plot_pairwise_features_vs_rmse(dataframe):
    features = dataframe.columns[:-1]  # Exclude the 'rmse' column

    plt.figure(figsize=(12, 8))
    sns.pairplot(dataframe, x_vars=features, y_vars='rmse', height=4, aspect=1.5)
    plt.suptitle('Pairwise Relationships between Features and RMSE')
    plt.show()


plot_pairwise_features_vs_rmse(dataset)
```
